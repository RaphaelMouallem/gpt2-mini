```python
import torch
import os
import gc
import os

from GPT_Config import GPTConfig 
from GPT import GPT as GPT
from builtTokenizer import TokenizerProcessor
from generateText import generate_text, chat_with_model

if torch.backends.mps.is_available():
    device = torch.device("mps")
    print("Using MPS (Apple Silicon GPU)")
elif torch.cuda.is_available():
    device = torch.device("cuda")
    print("Using CUDA")
else:
    device = torch.device("cpu")
    print("Using CPU")

print(f"Device: {device}")
```

```python
if device.type == 'mps':
    gc.collect()
    torch.mps.empty_cache()
    torch.backends.mps.enable_mps_fallback = True
    os.environ['PYTORCH_MPS_HIGH_WATERMARK_RATIO'] = '0.0'
    torch.mps.synchronize()
    print("MPS memory limit increased, cpu fallback initialised, cache cleaned")
```

```python
config = GPTConfig()
model = GPT(config, use_checkpointing=True)
state_dict = torch.load("...path/model.pth", map_location=device)
model.load_state_dict(state_dict, strict = True)
model.to(device)
```

```python
TOKENIZER_FILE_PATH = "...path/tokenizer.json"

tokenizer_processor = TokenizerProcessor(TOKENIZER_FILE_PATH)
```

```python
prompt = "Hi <|endoftext|>"

MAX_TOKENS = 200
TEMP = 0.7
TOP_K = 50

generated_text = generate_text(
    model, 
    tokenizer_processor,
    prompt=prompt, 
    max_new_tokens=MAX_TOKENS, 
    temperature=TEMP, 
    top_k=TOP_K,
    top_p=0.8,
    stop_token_ids = [50256]
)

print("\n" + "="*80)
print("PROMPT:")
print(prompt)
print("-" * 80)
print("GENERATED TEXT:")
print(generated_text)
print("="*80)
```
